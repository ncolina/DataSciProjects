# Scraping the bill data 
Information comes from http://www.congress.gov.ph/legisdocs/?v=bills

```python
import requests
from bs4 import BeautifulSoup

#congress = "http://www.congress.gov.ph/legisdocs/?v=bills"
#response=requests.get(congress)
#bills=response.content
file='untitled.txt'
data=open(file)
soup = BeautifulSoup(data,'lxml')

#print(soup.prettify())
```

```python
bill_names_raw=soup.select(".panel-body p")[::4]
bill_ids_raw=soup.select(".panel-heading .text-muted")
authors_raw=soup.select(".panel-body p")[1::4]
referral_raw=soup.select(".panel-body p")[2::4]
status_raw=soup.select(".panel-body p")[3::4]
```

```python
bill_names=[]
for name in bill_names_raw:
    bill_names.append(name.text)
bill_ids=[]
for ids in bill_ids_raw:
    bill_ids.append(ids.text)
authors=[]
for author in authors_raw:
    auth=author.text.split(";")
    auth[0]=auth[0].split(":")[1]
    for person in auth:
        person=person.replace(r'"(\w)*"','')
    authors.append(auth)
referral=[]
for ref in referral_raw:
    refer=ref.text.split(":")[1]
    referral.append(refer)
status=[]
for stat in status_raw:
    s=stat.text.split(":")[1]
    status.append(s)
```

```python
authors
```

```python
print(len(authors))
print(len(bill_names))
print(len(bill_ids))
```

```python
import pandas as pd
bill_data=pd.DataFrame()
```

```python
bill_data['bill_ids']=bill_ids
bill_data['bill_names']=bill_names
bill_data['authors']=authors
bill_data['referral']=referral
bill_data['status']=status
```

```python
bill_data.head()
bill_data.shape
```

```python
import numpy as np
auth=bill_data['authors'].apply(pd.Series)
#auth=auth.rename(columns = lambda x :"author_"+str(x))
auth.replace(' ', np.NaN,inplace=True)
auth=auth.dropna(axis=1, how='all')
auth.head()
```

```python
bill_data=pd.concat([bill_data[:],auth[:]],axis=1)
```

```python
d=pd.melt(bill_data, id_vars=['bill_ids','bill_names','authors','referral','status'])
d.sort_values(['bill_ids','variable'])
```

```python
data=pd.get_dummies(d,columns=['value'])
```

# Scraping the congressman data
data from https://en.wikipedia.org/wiki/17th_Congress_of_the_Philippines

```python
wiki = "https://en.wikipedia.org/wiki/17th_Congress_of_the_Philippines"
#header = {'User-Agent': 'Mozilla/5.0'} #Needed to prevent 403 error on Wikipedia
#req = requests.Request(wiki,headers=header)
page = requests.get(wiki)
soup = BeautifulSoup(page.content, "html.parser")
tables = soup.findAll("table", { "class" : "wikitable" })
```

```python
def extract_data(table):
    rows=table.findAll("tr")
    row_lengths=[len(r.findAll(['td','th'])) for r in rows]
    ncols=max(row_lengths)
    nrows=len(rows)
    data=[]
    for i in range(nrows):
        rowD=[]
        for j in range(ncols):
            rowD.append('')
        data.append(rowD)
    
    # process html
    for i in range(len(rows)):
        row=rows[i]
        rowD=[]
        cells = row.findAll(["td","th"])
        c=[]
        #for cell in cells:
        #    if cell.text !='':
        #        c.append(cell)
        #cells=c
        for j in range(len(cells)):
            if (len(cells) < ncols):
                cell=cells[j]
                
                #lots of cells span cols and rows so lets deal with that
                cspan=int(cell.get('colspan',1))
                rspan=int(cell.get('rowspan',1))
                for k in range(rspan):
                    for l in range(cspan):
                        data[i+k][j+l+1]+= cell.text
            else:
                cell=cells[j]
                
                #lots of cells span cols and rows so lets deal with that
                cspan=int(cell.get('colspan',1))
                rspan=int(cell.get('rowspan',1))
                for k in range(rspan):
                    for l in range(cspan):
                        data[i+k][j+l]+= cell.text
        
        #data.append(rowD)
    return pd.DataFrame(data[1:], columns=data[0])
```

```python
congressman_data=extract_data(tables[-2])
congressman_data.columns=[ 'Province/City', 'District','', 'Representative','RepresentativeParty', 'Term', 'Bloc']
congressman_data=congressman_data[['Representative', 'Province/City', 'District', 'RepresentativeParty', 'Term', 'Bloc']]
congressman_data['Representative']=congressman_data['Representative'].str.replace(r'[\[\]]|[0-9]', '')
congressman_data['RepresentativeParty']=congressman_data['RepresentativeParty'].str.replace(r'[\[\]]|[0-9]', '')
congressman_data.sort_values('Representative',inplace=True)
congressman_data.reset_index(drop=True,inplace=True)
congressman_data.head(10)
```

```python
data = data.T
```

```python
from sklearn.cluster import KMeans
kmeans= KMeans(n_clusters=5)
kmeans.fit(data.iloc[7:])
labels=kmeans.labels_
```

```python
kmeans.score(data.iloc[7:])
```

# Visualization

```python
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.decomposition import PCA
pca_2 = PCA(2)

# Turn the vote data into two columns with PCA
plot_columns = pca_2.fit_transform(data.iloc[7:])

# Plot senators based on the two dimensions, and shade by cluster label
# You can see the plot by clicking "plots" to the bottom right
plt.scatter(x=plot_columns[:,0], y=plot_columns[:,1], c=labels)
plt.show()
```
